```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.api as sm
import os
```

```python
dataFrame = pd.read_csv("Advertising.csv")
```

```python
dataFrame.head()
```

```python
dataFrame.describe()
```

```python
dataFrame[u'TV'].value_counts()
```

```python
dataFrame.drop(['Unnamed: 0'],axis=1)
```

```python
plt.figure(figsize = (10,8))
plt.scatter(dataFrame['TV'],dataFrame['sales'],c='purple')
plt.xlabel(" ($) Gastos em propaganda de TV")
plt.ylabel("($) Vendas")
plt.title('TV X Vendas')
plt.show()
```

```python
# Criar o modelo para prever o quanto teremos de retorno com TV

X = dataFrame['TV'].values.reshape(-1,1)
y = dataFrame['sales'].values.reshape(-1,1)


reg = LinearRegression()
reg.fit(X,y)

print('O modelo é: Vendas = {:.5} + {:.5} X'.format(reg.intercept_[0],reg.coef_[0][0]))
```

```python
f_previsoes = reg.predict(X)
plt.figure(figsize = (10,8))
plt.scatter(dataFrame['TV'],dataFrame['sales'],c='purple')
plt.plot(dataFrame['TV'],f_previsoes,c='green',linewidth=3,linestyle=':')
plt.xlabel(" ($) Gastos em propaganda de TV")
plt.ylabel("($) Vendas")
plt.title('TV X Vendas')
plt.show()
```
